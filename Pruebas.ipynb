Generar un listado de 1000 estudiantes:
- Verificar que no se repitan nombres ni apelllidos (personas totalmente distintas).

```python
dfn
```

```python
# .sample exttrajo el numero de datos aleatorios necesarios
## Return a random sample of items from an axis of object.

dfaa= dfa.sample(len(dfn)*2)

dfaa
```

```python
# .tolist() convierte una serie en una lista

nombres = dfn['name'].tolist()

genero = dfn['type'].tolist()

apellidos = dfaa['lastname'].tolist()

can= dfaa['cant'].tolist()

prov= dfaa['provincia_id'].tolist()

pn= dfaa['provincia_nombre'].tolist()

print(nombres)

print(apellidos)

print(genero)

print(can)

print(prov)

print(pn)
```

```python
#lista de primeros y segundos apellidos

primer= apellidos[:len(dfn)-1]

segundo= apellidos[len(dfn):len(dfn)*2]

g= genero[:len(dfn)]

c= can[:len(dfn)*2]

pi= prov[:len(dfn)*2]

p= pn[:len(dfn)*2]

print(primer)

print(segundo)

print(g)

print(c)

print(pi)

print(p)

```

```python
# La opción rn.sample elige una cantidad de datos aleatorios
nombres_apellidos = rn.sample(list(zip(nombres, primer, segundo, g, c, pi, p)), 1000)

print(nombres_apellidos)
```

```python
data = pd.DataFrame(nombres_apellidos, columns=['nombre', 'primer apellido','segundo apellido', 'Género', 'Cant', 'ID Provincia', 'Provincia'])

#Para trabajar los apellidos en mayusculas usamos lo siguiente:

data['primer apellido']= data['primer apellido'].str.upper()

data['segundo apellido']= data['segundo apellido'].str.upper()



print(data)
```
